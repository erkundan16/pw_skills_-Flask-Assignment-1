What is Flask Framework?

Flask is a micro web framework written in Python. It’s a Python module that lets you develop web applications easily. It has a small and easy-to-extend core. It doesn’t include an ORM (Object Relational Manager) or such features. It does have many cool features like URL routing, template engine. It is a WSGI web app framework.

Advantages :

1.Minimalistic Approach 
2.Easy to Understand
3.Flexibility and Extensibility
4.Testing Support
5.Control
6.Versatility
7.Affordable Development
8.Scalability

Q3. App Routing in Flask

App routing in Flask is a mechanism that maps specific URLs to functions in your application. These functions are called view functions and are responsible for handling requests to the application and returning responses.

We use app routes in Flask for the following reasons:

1.URL Handling

2.Organization

3.User Experience

4.Navigation Control

In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint with the method provided. Here’s a Python code to demonstrate the working of the url_for() function:


from flask import Flask, url_for
app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/welcome')
def welcome():
    return 'Welcome Page'

@app.route('/contact')
def contact():
    return 'Contact Page'

with app.test_request_context():
    print(url_for('home'))
    print(url_for('welcome'))
    print(url_for('contact'))
